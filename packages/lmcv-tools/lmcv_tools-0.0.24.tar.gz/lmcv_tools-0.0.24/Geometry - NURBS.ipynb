```python
from lmcv_tools.models.geometry import NURBS, BSpline
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = NURBS(
    2,
    [0, 0, 0, 1, 1, 1],
    [
        [0, 0],
        [1, 0],
        [2, 0],
    ],
    [1.0, 1.0, 1.0]
)

n.degree_elevation(1)
n.knot_insertions([0.25, 0.5, 0.5, 0.5, 0.75])
n.control_points[4][1] = 0.5
xp = [p[0] for p in n.control_points]
yp = [p[1] for p in n.control_points]
plt.plot(xp, yp, 'ob')

t = np.linspace(0, 1, 100)
x = [n(ti)[0] for ti in t]
y = [n(ti)[1] for ti in t]
plt.plot(x, y, '-r')

# plt.xlim(-80, 80)
plt.ylim(-0.1, 1.0)
plt.show()
```
